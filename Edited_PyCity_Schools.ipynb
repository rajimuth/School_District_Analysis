```python
# function used: pd.cut () - transform continuous data in to categorical data
# Use cut when you need to segment and sort data values into bins. 
# pd.cut()- this creates a new datafram or sweries on the given ranges
# Supports binning into an equal number of bins, or a pre-specified array of bins.
#  used to spearate the array element in to different bins
```

```python
# Cut the per_school_capita into the spending ranges.
# std dev is approximately 30
# not going with $578 as lowest bin because there is only one school below that range
#  4 schools spend less than $585
# each bin has a lower edge and upper edge
#  array that holds the spending bin ranges will have 5 values because when pandas create the lower edge it need a value lower than the the lowest value in the column of per_school_capita series
#  notice $ isnt includedwith ranges
# this is because numbers in the per_school_capita series are the float64 datatype
# error is $ sign is added
#  below list defines spending bins from bin1: 0- 585, Bin2: 586 -615, Bin3: 616 - 645, Bin4: 646:675
# Cut the per_school_capita into the spending ranges.
# spending_bins = [0, 585, 615, 645, 675]
# pd.cut(per_school_capita, spending_bins)
spending_bins = [0, 585, 615, 645, 675]
pd.cut(per_school_capita, spending_bins)
```

```python
# Looking at this output, it seems that we didn't group the schools fairly. 
# We need to adjust our ranges so that we have three or four schools in each range.
#  Let's increase the [585, 615] range to [585, 630] and change the third range to [630, 645]. 
# The spending_bins will look like this: 

# [0, 585, 615, 645, 675]
```

```python
# Edit bin values and Cut the per_school_capita into the spending ranges.
# # Cut the per_school_capita into the spending ranges.

spending_bins = [0, 585, 630, 645, 675]
per_school_capita.groupby(pd.cut(per_school_capita, spending_bins)).count()
```

```python
# Now let's name each range. 
# We can label the ranges using a list of string values. 
# Establish the spending bins and group names.
# Establish the spending bins and group names.
# spending_bins = [0, 585, 630, 645, 675]
# group_names = ["<$584", "$585-629", "$630-644", "$645-675"]
spending_bins = [0, 585, 630, 645, 675]
group_names = ["<$584", "$585-629", "$630-644", "$645-675"]
```

```python
#we have our correct spending bins and group names, 
# we can "cut" the per_school_capita Series to get the four spending bins
#  and add them to the per_school_summary_df DataFrame.
# create a new column in the per_school_summary_df dataframe
# values will be assigned the spending bins from the per_school_capita Series.
```

```python
# question: from module 4.11.1
# To the left of the equals sign, we add a new column called "Spending Ranges (Per Student)"{style="font-size: 1rem;"} to the per_school_summary_df DataFrame.
```

```python
# Categorize spending based on the bins.
# _df["new column name"] = pd.cut(per_school_capita, spending_bins, labels = group_names (from the list we created above))
per_school_summary_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels=group_names)
```
